```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("/content/istockphoto-911274308-612x612.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#plt.imshow(img)


kernel=[[1,2,1],
        [1,2,1],
       [1,2,1]]
custom_kernel=np.array(kernel,dtype=np.float32)/12
filtered_image = cv2.filter2D(img, -1, custom_kernel)
plt.imshow(filtered_image)
```
