**create a custom kernel and apply this image** . Enhances edges and details in an image.Increases contrast between different regions.Useful for highlighting object boundaries in image processing.




```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("/content/drive/MyDrive/COMPUTER VISION /download (1).jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#plt.imshow(img)


kernel=[[1,5,1],
        [1,5,1],
       [1,5,1]]
custom_kernel=np.array(kernel,dtype=np.float32)
filtered_image = cv2.filter2D(img, -1, custom_kernel)
plt.imshow(filtered_image)
```

**why use odd number in kernel** ? cause inn odd number there is a center point where other point are same distance from the centre. With odd-sized kernels, we can add equal padding on all sides

```python
img1=cv2.imread("/content/istockphoto-911274308-612x612.jpg")
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
#plt.imshow(img1)
kaernel1=np.ones((5,5),np.float32)/25
blur_img=cv2.filter2D(img1,-1,kaernel1)
#plt.imshow(blur_img)

kernel2=np.ones((15,15),np.float32)/225
blur_img2=cv2.filter2D(img1,-1,kernel2)
#plt.imshow(blur_img2)

plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.imshow(img1)
plt.subplot(1,3,2)
plt.imshow(blur_img)
plt.subplot(1,3,3)
plt.imshow(blur_img2)
plt.show()
```

```python
img2=cv2.imread("/content/istockphoto-911274308-612x612.jpg")
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)

kernel11=np.ones((20,20),np.float32)/400
blur_img11=cv2.filter2D(img2,-1,kernel11)
#plt.imshow(blur_img11)

sharpen_kernel=np.array([[0,-1,0],[-1,5,-1],[0,-1,0]],np.float32)
shap_img=cv2.filter2D(img2,-1,sharpen_kernel)
#plt.imshow(shap_img)

sharpen_kerne1 =np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]],np.float32)
shap_img=cv2.filter2D(img2,-1,sharpen_kerne1 )

sharpen_kernel2=np.array([[0,-.5,0],[-.5,3,-.5],[0,-.5,0]],np.float32)
shap_img=cv2.filter2D(img2,-1,sharpen_kernel2)




plt.figure(figsize=(15,5))
plt.subplot(1,4,1)
plt.title("blur image")
plt.imshow(blur_img11)
plt.subplot(1,4,2)
plt.title(" Standard Sharpening")
plt.imshow(shap_img)
plt.subplot(143)
plt.title("Strong Sharpening")
plt.imshow(shap_img)
plt.subplot(1,4,4)
plt.title("Very Strong Sharpening")
plt.imshow(shap_img)
plt.show()
```
