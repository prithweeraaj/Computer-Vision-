```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1=cv.imread("/content/drive/MyDrive/COMPUTER VISION /half lemon.jpg",cv.IMREAD_GRAYSCALE)
img2=cv.imread("/content/drive/MyDrive/COMPUTER VISION /lemon half 2.jpg",cv.IMREAD_GRAYSCALE)
img3=cv.imread("/content/drive/MyDrive/COMPUTER VISION /download.jpg",cv.IMREAD_GRAYSCALE)
img4=cv.imread("/content/drive/MyDrive/COMPUTER VISION /triumph-scrambler-400x-brand.jpg",cv.IMREAD_GRAYSCALE)
```

```python
#mean thresholding
T=np.mean(img1)
print(f"Mean Threshold value is {T}")
_,th1_img=cv.threshold(img1,T,255,cv.THRESH_BINARY)
plt.imshow(th1_img,cmap='gray')
```

```python
#fixed thresholding
print("the fixed thresholding is given below")
ret,th2_img=cv.threshold(img2,127,255,cv.THRESH_BINARY)
plt.imshow(th2_img,cmap='gray')
```

```python
# Otsu's Thresholding
print("the Otsu's thresholding is given below")
ret,th3_img=cv.threshold(img3,127,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
plt.imshow(th3_img,cmap='gray')
```

```python
#Triangle Thresholding
print("Triangle Thresholding")
T_triangle, th4_img1 = cv.threshold(img4,0,255, cv.THRESH_BINARY + cv.THRESH_TRIANGLE)
plt.imshow(th4_img1,cmap='gray')
```
