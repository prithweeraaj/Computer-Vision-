```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread("/content/drive/MyDrive/COMPUTER VISION /sudoku1.jpg",cv.IMREAD_GRAYSCALE)
plt.imshow(img,cmap='gray')
```

```python
sobelx=cv.Sobel(img,-1,0,1,3)
plt.imshow(sobelx,cmap='gray')
plt.title("sobelx")
plt.axis('off')
```

```python
sobely=cv.Sobel(img,-1,1,0,3)
plt.imshow(sobely,cmap='gray')
plt.title("sobely")
plt.axis('off')
```

```python
abs_sobelx=np.absolute(sobelx)
abs_sobely=np.absolute(sobely)
both=np.sqrt(sobelx**2+sobely**2)
abs_sobelx=cv.convertScaleAbs(abs_sobelx)
abs_sobely=cv.convertScaleAbs(abs_sobely)
both = both.astype(np.float32)
both=cv.convertScaleAbs(both)
plt.imshow(both, cmap='gray')
plt.title("Combined Gradient")
plt.axis('off')
plt.show()
```
