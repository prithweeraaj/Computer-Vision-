```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
path="/content/drive/MyDrive/COMPUTER VISION /berry-1.jpg"
img= cv2.imread(path,cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img,cmap='gray')
```

```python
img.shape
```

```python
img_bgr=cv2.imread(path)
img_gray=cv2.cvtColor(img_bgr,cv2.COLOR_BGR2RGB)
plt.imshow(img_bgr)
```

```python
adaptive_mean=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,3,2)
plt.imshow(adaptive_mean,cmap='gray')
```

```python
adaptive_mean=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,5,10)
plt.imshow(adaptive_mean,cmap='gray')
```

```python
ret1,ostu1=cv2.threshold(img,160,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
ret2,ostu2=cv2.threshold(img,255,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.subplot(1,2,1)
plt.imshow(ostu1,cmap='gray')
plt.subplot(1,2,2)
plt.imshow(ostu2,cmap='gray')

plt.show()
```
