```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img=cv.imread("/content/drive/MyDrive/COMPUTER VISION /20210329_180900.jpg",cv.IMREAD_GRAYSCALE)
plt.imshow(img,cmap="gray")
```

```python
sobelx=cv.Sobel(img,-1,1,0,3)
plt.imshow(sobelx,cmap="gray")
plt.title("sobelx")
plt.axis('off')
```

```python
sobely=cv.Sobel(img,-1,0,1,3)
plt.imshow(sobely,cmap="gray")
plt.title("sobely")
```

```python
sobelx=cv.Sobel(img,-1,1,0,3)
sobely=cv.Sobel(img,-1,0,1,3)
sobelx=cv.convertScaleAbs(sobelx)
sobely=cv.convertScaleAbs(sobely)
sobelxy=cv.addWeighted(sobelx,1,sobely,2,0)
plt.imshow(sobelxy,cmap="gray")
plt.title("sobelxy")
plt.axis('off')
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(141)
plt.imshow(img, cmap='gray')
plt.title("Horizontal Gradient (Sobel X)")
plt.axis('off')
plt.subplot(142)
plt.imshow(sobelx, cmap='gray')
plt.title("Horizontal Gradient (Sobel X)")
plt.axis('off')
plt.subplot(143)
plt.imshow(sobely, cmap='gray')
plt.title("Horizontal Gradient (Sobel X)")
plt.axis('off')
plt.subplot(144)
sobelx=cv.Sobel(img,-1,1,0,3)
sobely=cv.Sobel(img,-1,0,1,3)
sobelx=cv.convertScaleAbs(sobelx)
sobely=cv.convertScaleAbs(sobely)
sobelxy=cv.addWeighted(sobelx,1,sobely,2,0)
plt.imshow(sobelxy,cmap="gray")
plt.title("sobelxy")
plt.axis('off')
plt.show()
```
