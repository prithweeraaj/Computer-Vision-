```python
image2=cv2.imread("/content/drive/MyDrive/COMPUTER VISION /download.jpg")
cv2_imshow(image2)
```

```python
rotated_image=cv2.rotate(image2, cv2.ROTATE_180)
cv2_imshow(rotated_image)
```

```python
image2.shape
```

```python
#get the image dimensions:
(h,w)=image2.shape[:2]
center=(w//2,h//2)
#create a rotation martix.
M1=cv2.getRotationMatrix2D(center,30,1.0)
M2=cv2.getRotationMatrix2D(center,45,1.0)
M3=cv2.getRotationMatrix2D(center,-90,1.0)
m4=cv2.getRotationMatrix2D(center,180,2)
m5=cv2.getRotationMatrix2D(center,270,1.0)
m6=cv2.getRotationMatrix2D(center,90,1.0)
m7=cv2.getRotationMatrix2D(center,0,1.0)
m8=cv2.getRotationMatrix2D(center,120,1.0)
m9=cv2.getRotationMatrix2D(center,150,1.0)
m10=cv2.getRotationMatrix2D(center,180,1.0)



# rotate  the image
rotated_image_30=cv2.warpAffine(image2,M1,(w,h))
rotated_image_45=cv2.warpAffine(image2,M2,(w,h))
rotated_image_90=cv2.warpAffine(image2,M3,(w,h))
rotated_image_180=cv2.warpAffine(image2,m4,(w,h))
rotated_image_270=cv2.warpAffine(image2,m5,(w,h))
rotated_image_90=cv2.warpAffine(image2,m6,(w,h))
rotated_image_0=cv2.warpAffine(image2,m7,(w,h))
rotated_image_120=cv2.warpAffine(image2,m8,(w,h))
rotated_image_150=cv2.warpAffine(image2,m9,(w,h))
rotated_image_180=cv2.warpAffine(image2,m10,(w,h))

# display
plt.figure(figsize=(50,15))

plt.subplot(251)
plt.imshow(rotated_image_30)
plt.title("rotated_image_30")


plt.subplot(2,5,2)
plt.imshow(rotated_image_45)
plt.title("rotated_image_45")


plt.subplot(2,5,3)
plt.imshow(rotated_image_90)
plt.title("rotated_image_90")

plt.subplot(2,5,4)
plt.imshow(rotated_image_180)
plt.title("rotated_image_180")

plt.subplot(2,5,5)
plt.imshow(rotated_image_270)
plt.title("rotated_image_270")

plt.subplot(2,5,6)
plt.imshow(rotated_image_90)
plt.title("rotated_image_90")

plt.subplot(2,5,7)
plt.imshow(rotated_image_0)
plt.title("rotated_image_0")

plt.subplot(2,5,8)
plt.imshow(rotated_image_120)
plt.title("rotated_image_120")

plt.subplot(2,5,9)
plt.imshow(rotated_image_150)
plt.title("rotated_image_150")

plt.subplot(2,5,10)
plt.imshow(rotated_image_180)
plt.title("rotated_image_180")



plt.tight_layout()
plt.show()
```

4. Plot a 2x2 grid with original, three resized images with different
interpolation method


```python
image3=cv2.imread("/content/drive/MyDrive/COMPUTER VISION /c.jpg")
plt.imshow(image3)
```

```python
image3 = cv2.cvtColor(image3, cv2.COLOR_BGR2RGB)
plt.imshow(image3)
```

```python


resized_nearest=cv2.resize(image3,(200,200),interpolation=cv2.INTER_NEAREST)
resized_linear=cv2.resize(image3,(200,200),interpolation=cv2.INTER_LINEAR)
resized_cubic=cv2.resize(image3,(200,200),interpolation=cv2.INTER_CUBIC)

plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
plt.imshow(resized_nearest)
plt.title("resized_nearest")

plt.subplot(2,2,2)
plt.imshow(resized_linear)
plt.title("resized_linear")

plt.subplot(2,2,3)
plt.imshow(resized_cubic)
plt.title("resized_cubic")

plt.subplot(2,2,4)
plt.imshow(image3)
plt.title("original")

plt.tight_layout()
plt.show()
```
