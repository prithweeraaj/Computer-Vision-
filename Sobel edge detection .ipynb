```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread("/content/drive/MyDrive/COMPUTER VISION /sudoku1.jpg",cv.IMREAD_GRAYSCALE)
plt.imshow(img,cmap='gray')
```

```python
sobelx=cv.Sobel(img,-1,0,1,3)
plt.imshow(sobelx,cmap='gray')
plt.title("sobelx")
plt.axis('off')
```

```python
sobely=cv.Sobel(img,-1,1,0,3)
plt.imshow(sobely,cmap='gray')
plt.title("sobely")
plt.axis('off')
```

```python
sobely=cv.Sobel(img,-1,0,1,3)
sobely=cv.Sobel(img,-1,1,0,3)
sobelx=cv.convertScaleAbs(sobelx)
sobely=cv.convertScaleAbs(sobely)
sobelxy=cv.addWeighted(sobelx,1,sobely,1,0)
plt.imshow(sobelxy,cmap='gray')
plt.title("dx and dy combained ")
plt.axis('off')
```
