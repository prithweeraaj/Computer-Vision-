```python
import cv2 as cv
import numpy as np
from matplotlib import  pyplot as plt
```

```python
img=cv.imread('/content/drive/MyDrive/COMPUTER VISION /istockphoto-185284489-612x612.jpg',cv.IMREAD_GRAYSCALE)
#img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img,cmap='gray')#coverrting the gray scale
```

```python
#GOlbal thresholding
ret, th1=cv.threshold(img,172,255,cv.THRESH_BINARY)
plt.imshow(th1,cmap='gray')
print(ret)


```

```python
ret,th2=cv.threshold(img,127,255,cv.THRESH_BINARY_INV)
plt.imshow(th2,cmap='gray')
print(ret)
```

```python
ret,th2=cv.threshold(img,230,255,cv.THRESH_BINARY_INV)
plt.imshow(th2,cmap='gray')
print(ret)
```

```python
ret,th2=cv.threshold(img,220,255,cv.THRESH_TRUNC)
plt.imshow(th2,cmap='gray')
print(ret)
```

```python
ret,thresh4 = cv.threshold(img,150,255,cv.THRESH_TOZERO)
plt.imshow(thresh4,cmap='gray')
print(ret)
```

```python
ret,thresh5 = cv.threshold(img,127,255,cv.THRESH_TOZERO_INV)
plt.imshow(thresh5,cmap='gray')
print(ret)
```
